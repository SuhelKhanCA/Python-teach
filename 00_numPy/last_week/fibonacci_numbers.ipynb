```python
zeroth = 0
first = 1
print(zeroth)
print(first)

for _ in range(1, 6):
    next = zeroth + first
    zeroth = first
    first = next

print(next)
```

```python
def fib(n):
  if n == 1 or n == 0:
    return n

  return fib(n-1) + fib(n-2)


for i in range(1, 9):
  print(fib(i))
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1, 2, 5, 2, 9, 23])
y = np.array([0, 11, 15, 20, 12, 4])
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.title("Ye title hai")
plt.show()
```

```python
plt.plot(x, y)
plt.title("Ye title hai")
plt.xlabel("X axis hai")
plt.show()
```

```python
plt.plot(x, y)
plt.title("Ye title hai")
plt.xlabel("X axis hai")
plt.ylabel("Y axis hai")
plt.show()
```
